```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import datasets
from sklearn import preprocessing
from sklearn import decomposition
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import seaborn as sns
import matplotlib.pyplot as plt
```

# Part A (PCA)

```python
np.random.seed(5)
df = pd.read_csv('iris_data.txt',names=['sepal length','sepal width','petal length','petal width','species'])
le = preprocessing.LabelEncoder()
target = le.fit(df['species'])
target = le.transform(df['species'])
centers = [[1, 1], [-1, -1], [1, -1]]
```

```python
df1 = df[['sepal length','sepal width','petal length','petal width']]
X = df1.values
```

```python
target_names = df.species.unique()
```

```python
pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)
```

```python
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
```

```python
for color, i, target_name in zip(colors, [0, 1, 2],target_names):
    plt.scatter(X_r[target == i, 0], X_r[target == i, 1], color=color, alpha=.8, lw=lw,
                label=target_names[i])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
```

# Part B (TSNE)

```python
from sklearn.manifold import TSNE
```

```python
X_tsne = TSNE(n_components=2).fit_transform(X)
```

```python
for color, i, target_name in zip(colors, [0, 1, 2],target_names):
    plt.scatter(X_tsne[target == i, 0], X_tsne[target == i, 1], color=color, alpha=.8, lw=lw,
                label=target_names[i])
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('t-SNE of IRIS dataset')
```

# Part C

```python
swiss,target_swiss = datasets.make_swiss_roll(n_samples=2000, noise=0.0, random_state=None)
```

## PCA on swiss

```python
pca = PCA(n_components=2)
X_pca_swiss = pca.fit(swiss).transform(swiss)
plt.scatter(X_pca_swiss[:,0],X_pca_swiss[:,1],cmap='rainbow',c=target_swiss)
```

# TSNE on SWISS Dataset

```python
X_tsne_swiss = TSNE(n_components=2).fit_transform(swiss)
```

```python
plt.scatter(X_tsne_swiss[:,0],X_tsne_swiss[:,1],cmap='rainbow',c=target_swiss)
```
